### Lets Preprocess the brands and models column

```python
#Unique models in the dataset
models = df['model'].unique()
models
```

```python
#Function to clean model names
#This function will replace common abbreviations and typos in the model names with their full forms
def clean_model_names(df, column='model'):
    replacements = {
        'vehicl': 'vehicle',
        'pk': 'pickup',
        'dr': 'door',
        'drw': 'dual rear wheel',
        'srw': 'single rear wheel',
        'ld': 'limited',
        'd': 'drive',
        'door': 'doors',  #unify
        'el': 'EL',
        'trail': 'Trailblazer',  #or Trail
        'country': 'Town & Country',
        'max': 'C-Max',
        'note': 'Note',  #assuming Nissan Note
        'pioneer': 'Pioneer',  #assuming valid model
        'sundance': 'Sundance',
        'energi': 'Energi',
    }
    df[column] = df[column].str.lower()
    df[column] = df[column].replace(replacements)
    return df
```

```python
df = clean_model_names(df, column='model')
```

```python
#Uppercasing the brands and model naems that are 3 characters or less
#This is a common practice to standardize the data and make it easier to analyze
df.loc[:, 'brand'] = df['brand'].apply(lambda x: x.upper() if isinstance(x, str) and len(x) <= 3 else x)
df.loc[:, 'model'] = df['model'].apply(lambda x: x.upper() if isinstance(x, str) and len(x) <= 3 else x)
```

```python
#Applying title case to all string columns
df = df.apply(lambda col: col.str.title() if col.dtype == 'object' else col)
df['country'] = df['country'].str.upper()
```

### Now, lets preprocess the condition column

```python
#renaming the column name to time_left
df.rename(columns={'condition': 'time_left'})
```

```python
#unique values in the condition column
df["condition"].unique()
```

```python
#Function to clean the time_left column w.r.t Days, Hours and Minutes
def extract_time_components(row):
    import re

    # Default values
    days = hours = minutes = 0

    if 'Expired' in row:
        return pd.Series([0, 0, 0])

    # Extract using regex
    day_match = re.search(r'(\d+)\s*Days?', row, re.IGNORECASE)
    hour_match = re.search(r'(\d+)\s*Hours?', row, re.IGNORECASE)
    minute_match = re.search(r'(\d+)\s*Minutes?', row, re.IGNORECASE)

    if day_match:
        days = int(day_match.group(1))
    if hour_match:
        hours = int(hour_match.group(1))
    if minute_match:
        minutes = int(minute_match.group(1))

    return pd.Series([days, hours, minutes])
```

```python
#Apply to DataFrame
df[['time_left_days', 'time_left_hours', 'time_left_minutes']] = df['condition'].apply(extract_time_components)
```

```python
#Convert to total hours and minutes
df['total_hours'] = df['time_left_days'] * 24 + df['time_left_hours'] + df['time_left_minutes'] / 60
df['total_minutes'] = df['time_left_days'] * 24 * 60 + df['time_left_hours'] * 60 + df['time_left_minutes']
```

```python
#Preview of the dataset after cleaning the time_left column
df.head(5)
```

```python
#Drop the repeated and unnecessary columns
df.drop(columns=['condition', 'time_left_days','time_left_hours', 'time_left_minutes','total_minutes'], inplace=True)
df.head(10)
```

```python
#Rename the total_hours column to time_left_hours for better understanding
df.rename(columns={'total_hours': 'time_left_hours'}, inplace=True)
```

```python
#Title case for all columns
df.columns = [col.title() for col in df.columns]
df.head(10)
```
